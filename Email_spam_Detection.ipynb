Preprocessing of the dataset is illustrated below

```python
from nltk.corpus import stopwords
import string
from nltk.stem import PorterStemmer
ps = PorterStemmer()
def processed_text(text):
  text = nltk.word_tokenize(text.lower())
  s = []
  for i in text:
    if i.isalnum():
      s.append(i)
      text = s[:]
      s.clear()
      for i in text:
        if i not in stopwords.words('english') and i not in string.punctuation:
          stemming = ps.stem(i)
          s.append(stemming)
        return " ".join(s)
```

```python
processed_text('My name is oluwaseyi and i am a very fine boy')
```

```python
clean_data['New_Text'] = clean_data['Email Text'].apply(processed_text)
clean_data.head()
```

BUILDING A MODEL 


```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
tf = TfidfVectorizer(max_features= 3000)
X = tf.fit_transform(clean_data["New_Text"]).toarray()
#distributes it into an array
X
```

```python
y = np.array(clean_data['Class'])
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state= 48, test_size=0.3)
X_train.shape
```

```python
y_test.shape
```

```python
from sklearn.svm import LinearSVC
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix, roc_auc_score


# Create an SVM classifier
clf = svm.SVC(kernel='linear')

# Train the classifier on the training data
clf.fit(X_train, y_train)

# Make predictions on the test data
y_pred = clf.predict(X_test)

# Evaluate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

```

```python
class_report = classification_report(y_test, y_pred)
print("Classification report:" ,class_report )
```

```python
 import pandas as pd
 import numpy as np
 import nltk

 df = pd.read_csv("/content/reviews_for_classification.csv")
```

```python
df
```

```python


# Tokenize the 'Text' column
df['review_head'] = df['review_head'].apply(lambda x: nltk.word_tokenize(x))

# Print the tokenized data
print(df['review_head'])

# Perform POS tagging
df['pos_tags'] = df['review_head'].apply(lambda x: nltk.pos_tag(x))

# Print the POS tagged data
print(df['pos_tags'])


```
